```python
final_df = []
counter = 1

for file in tqdm(filenames):
    with open(file, 'r') as f:
        
        temp_df = pd.json_normalize(safe_load(f))
        temp_df['match_id'] = counter
        final_df.append(temp_df)
        
        counter += 1

        if counter == 2000:
            break


len(final_df)
```

```python
final_df = pd.DataFrame()

match_id = 1

for index, row in df.iterrows():
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    city = []
    venue = []
    toss_decision = []
    toss_winner = []

    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            teams.append(row['info.teams'])
            batting_team.append(row['innings'][0]['1st innings']['team'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            toss_decision.append(row['info.toss.decision'])
            toss_winner.append(row['info.toss.winner'])

            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_of_dismissed.append(0)

    temp_df = pd.DataFrame({
        'match_id': match_id,
        'teams': teams,
        'toss_winner': toss_winner,
        'toss_decision': toss_decision,
        'batting_team': batting_team,
        'ball': ball_of_match,
        'batsman': batsman,
        'bowler': bowler,
        'runs': runs,
        'player_dismissed': player_of_dismissed,
        'city': city,
        'venue': venue
    })

    final_df = pd.concat([final_df, temp_df], ignore_index=True)
    match_id += 1
```

```python
def bowl(row):
    for team in row['teams']:
        if team not in row['batting_team']:
            return team

final_df['bowling_team'] = final_df.apply(bowl, axis=1)

```

```python
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
df[df['match_id'] == 1]
```

```python
def distribute_over_ball(row):
    overs = row['ball'].astype('float').apply(lambda x: str(x).split('.')[0])
    balls = row['ball'].astype('float').apply(lambda x: str(x).split('.')[1])

    return overs, balls
    
distribute_over_ball(df)
```

```python
overs, balls = distribute_over_ball(df)

df['overs'] = overs
df['balls'] = balls
```

```python
df['overs'] = df['overs'].astype(int)
df['balls'] = df['balls'].astype(int)
```

```python
df.head(10)
```

```python
df.info()
```

```python
df['total_balls'] = (df['overs'] * 6) + df['balls']
```

```python
df.head()
```

```python
df.info()
```

```python
df['balls_left'] = 120 - df['total_balls']
```

```python
df['balls_left'] = df['balls_left'].apply(lambda x: 0 if x < 0 else x)
```

```python
df.head()
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x: 0 if x == 0 else 1)
```

```python
df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()
```

```python
df[df['match_id'] == 1]
```

```python
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df['run_rate'] = (df['current_score'] * 6) / df['total_balls']
```

```python
# Check for NaN values in the entire DataFrame
print(df.isna().sum())
```

```python
groups = df.groupby('match_id')
```

```python
groups
```

```python
match_ids = df['match_id'].unique()
```

```python
match_ids
```

```python
last_five = []

for id in match_ids:
    group_df = groups.get_group(id)  # Get DataFrame for each group
    numeric_cols = group_df.select_dtypes(include=['number'])  # Select numeric columns
    rolling_sum = numeric_cols.rolling(window=30).sum()  # Compute rolling sum
    last_five.extend(rolling_sum['runs'].values.tolist())  # Keep NaNs as well
```

```python
len(last_five)
```

```python
last_five
```

```python
df['last_five'] = last_five
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df, on='match_id')
```

```python
final_df = final_df[['batting_team', 'bowling_team', 'city', 'current_score', 'balls_left', 'wickets_left', 'run_rate', 'last_five', 'runs_x']]
```

```python
# working on ML algorithm
```

```python
X = final_df.drop(columns = ['runs_x'])
y = final_df['runs_x']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train.head()
```

```python
X_test
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
trf = ColumnTransformer(transformers=[
    ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city']),
], remainder='passthrough')
```

```python
trf
```

```python
pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', StandardScaler()),
    ('step3', XGBRegressor(n_estimators=1000, learning_rate=0.3, max_depth=12, random_state=1))
    # ('step3', RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=1, verbose=2)) # not giving great accuracy
    
])
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
print(r2_score(y_pred, y_test))
```

```python
print(mean_absolute_error(y_pred, y_test))
```
